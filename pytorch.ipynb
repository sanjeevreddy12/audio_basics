```python
import torch
import torch.nn as nn

# Example: 1 second of audio at 16 kHz
audio = torch.randn(16000)

print("Audio shape:", audio.shape)
print("Type:", audio.dtype)


class AudioModel(nn.Module):
    def __init__(self):
        super().__init__()
        # Linear layer: input 16000 features -> output 128 features
        self.linear = nn.Linear(16000, 128)

    def forward(self, x):
        # Add batch dimension if missing
        if x.dim() == 1:
            x = x.unsqueeze(0)  # (16000,) -> (1,16000)

        out = self.linear(x)
        return out


# Create model
model = AudioModel()

# Forward pass
output = model(audio)

print("Output shape:", output.shape)
print(output)
```
